```python
import numpy as np
import matplotlib.pyplot as plt

import requests
import json
import cv2

from PIL import Image 
from io import BytesIO
```

```python
def download(pano_id: str, **kwargs):
    content =  requests.get('https://streetviewpixels-pa.googleapis.com/v1/tile', params={'cb_client': 'apiv3', 'panoid': pano_id, **kwargs}).content 
    return Image.open(BytesIO(content))
```

```python
with open('../data/locs.json', 'r') as f:
    data = json.loads(f.read())['customCoordinates']
```

```python
for loc in [data[0]]:
    image = download(loc['panoId'])
    year = extract_year(image)

    display(image, year)
```

```python
from __future__ import annotations
import sys, re
from pathlib import Path

import cv2, pytesseract
import numpy as np

# ---------- constants --------------------------------------------------------
YEAR_RE  = re.compile(r"^(\d{4})$")
RANGE    = range(2008, 2026)          # inclusive upper bound
TESS_CFG = (
    "--psm 6 --oem 3 "
    "-c tessedit_char_whitelist='0123456789Ggoogle '"
)
# -----------------------------------------------------------------------------

def preprocess(img: np.ndarray) -> np.ndarray:
    return img


def extract_year(img: Image) -> int | None:
    img = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)
    pre = preprocess(img)

    plt.imshow(img)

    # full-frame OCR (word-level data)
    data = pytesseract.image_to_data(
        pre, config=TESS_CFG, output_type=pytesseract.Output.DICT
    )

    candidates: list[tuple[int, float]] = []

    for i, word in enumerate(data["text"]):
        if YEAR_RE.match(word):
            word = word[1:]
            return int(word)
    return None
```
